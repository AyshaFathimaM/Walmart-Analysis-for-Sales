## Initial Insights

Analyzing Walmart's sales data from the year 2019. Here are some key points from our initial analysis:

1. Dataset Overview: The dataset consists of 1,000 rows and 17 columns, organized by invoice ID.
2. Column Removal: The invoice ID column has been dropped as it is not needed for the analysis.
3. Data Types: Most columns have the correct data type; however, the date and time column is currently an object type and needs to be converted for better insights.
4. Data Quality: There are no missing or duplicate values in the dataset, so no null values need to be addressed.
5. Branch Performance: Branch A generated the most invoices, with 340, followed by Branch B with 332, and Branch C with 328.
6. Loyalty Programs: Customers enrolled in loyalty programs created more invoices (501) compared to non-members (499). Further analysis of the value of these invoices is recommended.
7. Popular Products: The most purchased product line was Fashion Accessories, followed by Food and Beverages, and Electronic Accessories.
8. Payment Methods: The preferred payment method is E-wallet, followed by Cash and Credit Card.
9. Product Pricing: The average unit price of products is $55.60, with a minimum price of $26.49 and a maximum of $99.96.
10. Quantity Purchased: On average, customers bought 5.51 units per transaction, with a minimum of 1 and a maximum of 10 units.

```python
# Feature engineering

#converting date column which is in object type into datetime type and extract year, month and day
```

```python
df.head()
```

```python
df.Date= pd.to_datetime(df.Date)
df.Date
```

```python
#Year month and day
```

```python
df.Date.dt.year
```

```python
df.Date.dt.month
```

```python
df.Date.dt.day
```

```python
df["Year"]= df.Date.dt.year
df["Month"]= df.Date.dt.month
df["Day"]= df.Date.dt.day
df.head()
```

```python
df["Month_name"]=df.Date.dt.month_name()
df["Day_name"] = df.Date.dt.day_name()
```

```python
df.head()
```

```python
df.info()
```

```python
df['Time_of_day'] = pd.cut(pd.to_datetime(df.Time).dt.hour,
                          bins=[5,12,16,23],
                          labels=['Day','Afternoon','Evening'],
                          include_lowest=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Month"], axis=1, inplace=True)
df.drop(["Day"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.Year.value_counts()
```

## Summary of Feature Engineering
1. Feature Engineering: Some feature engineering was performed on the dataset for improved analysis.
2. Change Data Type: The data type of the Date and Time columns was changed from object to datetime. This allows for the extraction of the month, day of the week, and time of day, which can provide useful insights.
3. Extract Month and Day Names: The month_name and day_name functions were used to determine the names of the months and days.
4. Categorize Time of Day: The time of day was categorized into morning, afternoon, and evening.
5. Drop Unnecessary Columns: Unnecessary columns were removed to avoid cluttering the dataset.
